```python
N_test = 100
rare_test = RareEvents_test(level = 8, p_0 = 0.1)
var_rel = []
for i in range(100):
    var_rel = np.append(var_rel,rare_test.adaptive_levels(N = N_test,status_tracking=False,descent_step = 0.1,reject_rate = 0.5)[1]*10.)
```

```python
rare_test = RareEvents_test(level = 8, p_0 = 0.75)
var_rel_big = []
for i in range(100):
    var_rel_big = np.append(var_rel_big,rare_test.adaptive_levels(N = N_test,status_tracking=False,descent_step = 0.1,reject_rate = 0.5)[1]*10.)
```

```python
plt.figure(figsize = [15,5])
sns.distplot(var_rel, color = 'darkred',label = 'p_0 = 0.1')
sns.distplot(var_rel_big, color = 'darkgreen',label = 'p_0 = 0.75')
plt.legend()
plt.title("Comparison for different $P_0$ ")
```

```python
print ("p_0 =0.1 :" + str(np.std(var_rel)))
print ("p_0 =0.75 :" + str(np.std(var_rel_big)))
```

```python
print ("p_0 =0.1 :" + str(np.std(var_rel*(var_rel<250))))
print ("p_0 =0.75 :" + str(np.std(var_rel_big*(var_rel_big<250))))
print (np.sum((var_rel<250)))
```

```python
print (np.sum((var_rel>0)))
```

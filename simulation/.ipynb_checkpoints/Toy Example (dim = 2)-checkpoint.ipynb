```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
```

# Toy Example (dim = 2)
* $X \sim \mathcal{N}(0,Id_2)$
* $S:\mathbb{R}^2 \to \mathbb{R}$ score function / blackbox i.e. we could simulate $S(X)$, but we don't know the form of $S$. here we take $S(X) = \mathrm{Norm}_{euclidien}(X) $
* Goal : estimate $\mathbb{P}(S(X)>q) < 10^{-6}$ (rare event)

```python
# simulation of X
num_X = 1000 #number of sample
X = np.random.normal(0,1,num_X*2).reshape(num_X,2)
def S(X):
    '''score function which is a black box'''
    if np.shape(X) == (2,):
        return np.linalg.norm(X)
    else:
        return np.linalg.norm(X, axis=1)

S(X[0,])# test
```

```python
plt.rcParams['figure.figsize'] = (15.0, 6.0)

plt.subplot(1,2,1)
plt.plot(X[:,0],X[:,1],"p")
plt.title("Illustration of X")

plt.subplot(1,2,2)
plt.hist2d(X[:,0],X[:,1],cmap = plt.cm.Blues,bins = 20)
plt.title("Illustration of X")
plt.colorbar()

```
